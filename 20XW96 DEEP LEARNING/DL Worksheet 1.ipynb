```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[1,1],[3,2]])
B = np.array([6,16])
```

```python
A_inv = np.linalg.inv(A)
```

```python
result = np.dot(A_inv, B)
```

```python
x,y = result
```

```python
x_vals = np.linspace(0, 8, 100)
y1_vals = 6 - x_vals
y2_vals = (16 - 3 * x_vals)/2

plt.plot(x_vals, y1_vals, label=r"$x + y = 6$")
plt.plot(x_vals, y2_vals, label=r"$3x + 2y = 16$")

intersection_x = 4
intersection_y = 2
plt.scatter(intersection_x, intersection_y, color='red', label="Intersection (4, 2)")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of Equations")
plt.grid(True)
plt.legend()

plt.show()
```

```python
n = 10  
p = 0.25 
mu = 100 
sigma = 15 
scale = 2.5  
sample_sizes = [100, 500, 1000]

for sample_size in sample_sizes:

    data_binomial = np.random.binomial(n, p, sample_size)
    data_normal = np.random.normal(mu, sigma, sample_size)
    data_standard_normal = np.random.standard_normal(sample_size)
    data_exponential = np.random.exponential(scale, sample_size)

    mle_binomial = np.mean(data_binomial) / n
    mle_mu_normal = np.mean(data_normal)
    mle_sigma_normal = np.std(data_normal)
    mle_standard_normal = np.mean(data_standard_normal)
    mle_exponential = 1 / np.mean(data_exponential)

    print(f"Sample size {sample_size}:")
    print(f"Binomial MLE: {mle_binomial:.4f}")
    print(f"Normal MLE (mu): {mle_mu_normal:.4f}")
    print(f"Normal MLE (sigma): {mle_sigma_normal:.4f}")
    print(f"Standard Normal MLE: {mle_standard_normal:.4f}")
    print(f"Exponential MLE: {mle_exponential:.4f}\n")
```
