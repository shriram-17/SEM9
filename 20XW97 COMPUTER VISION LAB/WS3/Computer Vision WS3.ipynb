```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
x = np.array([1,2,3])
h = np.array([1,-1])
```

```python
y = signal.convolve(x,h)
y
```

```python
t = np.linspace(0,1,100)
x = signal.square(2*np.pi * 5 *t)
#h  = np.array([1,-1]) #High Pass Fliter
h = 1/101*np.ones(11)
y  = signal.convolve(x,h,'same')
```

```python
plt.subplot(2,1,1),plt.plot(t,x),plt.title("Input Signal")
plt.subplot(2,1,2),plt.plot(t,y),plt.title("Output Signal")
plt.tight_layout()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = cv2.imread('cameraman.tif')
h = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]]) # Low pass fitler?
y = cv2.filter2D(x, -1, h)
plt.subplot(1, 2, 1), plt.imshow(x, cmap='gray')
plt.subplot(1, 2, 2), plt.imshow(y, cmap='gray')
```
