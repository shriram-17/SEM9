```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.zeros([256,256])
plt.imshow(arr,cmap='gray',vmin=0,vmax=255)
```

```python
arr[128,:] = 255
arr[:,128] = 255
```

```python
for i in range(0,len(arr)):
    for j in range(0,len(arr)):
        if i == j or (i == 256 - j):
            arr[i,j] = 255
```

```python
plt.imshow(arr,cmap='gray',vmin=0,vmax=255)
```

```python
circle = np.zeros([256,256])
```

```python
r = 20
for i in range(0,len(circle)):
    for j in range(0,len(arr)):
        if (i-128) ** 2 + (j-128) ** 2 < r ** 2:
            circle[i,j] = 255 
```

```python
plt.plot(circle[128,:])
```

```python
plt.imshow(circle,cmap='gray',vmin=0,vmax=255)
```
