### Quantization of the Signal 

#### To implement scalar quantifer
 
- Analog Signal -> Sampling -> Quantization -> DS
- Mapping larger set of values to smaller set of values

- Adavantage:
              - Compression
- Disadvantage:
              -  loss of information/data 

### Terminalogy
- DR : Dynamic Range
- Bit Budget : 'b'
- L = 2^ b
- S = DR/L

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t  = np.linspace(0,1,100)
x = np.sin(2*np.pi*5*t)
```

```python
plt.plot(t,x)
```

### Qunatization Problem

```python
for i in range(1,9):
    print(f"Bit Budget Size is {i}")
    dr = np.max(x) - np.min(x)
    b = i
    l = 2** b
    q = dr/l
    y = np.floor(x/q)*q
    plt.plot(t,x,'r',t,y,'k')
    plt.legend(['input','output'],loc = 1)
    plt.show()
```
