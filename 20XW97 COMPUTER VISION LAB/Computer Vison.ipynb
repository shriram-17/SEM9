```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img  = np.zeros([8,8])
```

## Method 1 Using the Euler Function

```python
m = len(img)
n = len(img[0]) 
```

```python
for i in range(0,m):
    for j in range(0,n):
        img[i,j] = (-1)**(i+j)
```

```python
plt.imshow(img,cmap="gray")
```

## Method 2 using modulo function

```python
img1 = np.zeros([8,8])
```

```python
m,n = len(img1),len(img1[0])
```

```python
for i in range(0,m):
    for j in range(0,n):
        img1[i,j] = (i+j) % 2
```

```python
plt.imshow(img,cmap="gray")
```

```python
for i in range(0,m):
    for j in range(0,n):
        img1[i,j] = (i+j) % 2
```

```python
img3 = np.zeros([200,200])
```

```python
for i in range(0,len(img3)):
    for j in range(0,len(img3[0])):
        if j % 2 == 0:
            img3[i,j] == 0
        else:
            img3[i,j] = 255
```

```python
plt.imshow(img3,cmap="gray")
```

```python
x = np.linspace(-128,128,256)
y = np.linspace(-128,128,256)
X,Y = np.meshgrid(x,y)
```

```python
fx = 10
fy = 10
```

```python
img_sin = np.sin(2 * np.pi * (fx * X + fy * Y))
```

```python
IMG = np.fft.fft2(img_sin)
IMG = np.fft.fftshift(IMG)
plt.subplot(1,2,1) 
plt.imshow(img_sin,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(np.abs(IMG),cmap="gray")
```
